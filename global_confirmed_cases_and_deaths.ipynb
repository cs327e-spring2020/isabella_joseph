```python
%%bigquery
select countries_and_territories, confirmed_cases, deaths, (deaths*100/confirmed_cases) as case_fatality_rate from covid19_data_analysis.covid_19_global_distribution_worldwide
where date = '2020-05-09' and confirmed_cases > 1000
order by case_fatality_rate DESC
limit 20
```

```python
%%bigquery
create table covid19_data_analysis.confirmed_cases_by_country as
select country_region, sum(_5_8_20) as confirmed_cases from covid19_data_analysis.jhu_confirmed_cases
group by country_region
order by confirmed_cases DESC
```

```python
%%bigquery
create table covid19_data_analysis.deaths_by_country as
select country_region, sum(_5_8_20) as deaths from covid19_data_analysis.jhu_deaths
group by country_region
order by deaths DESC
```

```python
%%bigquery
create table covid19_data_analysis.confirmed_cases_and_deaths_by_country as
SELECT c.country_region, c.confirmed_cases, d.deaths FROM covid19_data_analysis.confirmed_cases_by_country c
inner join covid19_data_analysis.deaths_by_country d on c.country_region = d.country_region
order by c.confirmed_cases desc
limit 10
```

```python
%%bigquery
select country_region, (deaths*100/confirmed_cases) as case_fatality_rate from covid19_data_analysis.confirmed_cases_and_deaths_by_country
where confirmed_cases > 1000
order by case_fatality_rate desc
limit 10
```

```python
%%bigquery
create or replace view covid19_data_analysis.v_confirmed_cases_and_deaths_by_country as
select country_region, confirmed_cases, deaths, (deaths*100/confirmed_cases) as case_fatality_rate from `sunny-advantage-266802.covid19_data_analysis.confirmed_cases_and_deaths_by_country`
where confirmed_cases > 1000
order by case_fatality_rate desc
```
